```python
#Reorder Linked List
#Input: head = [1,2,3,4]
#Output: [1,4,2,3]

class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1 

    def printList(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next

    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode
        temp = self.head 
        while temp.next is not None:
            temp = temp.next 
        temp.next = newnode
        self.length += 1

    def reorderList(self):
        slow,fast = self.head,self.head.next
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next 
        second = slow.next
        prev = slow.next = None
        while second is not None:
            temp = second.next
            second.next = prev 
            prev = second 
            second = temp 
        
        first,second = self.head,prev 
        while second is not None:
            temp1,temp2 = first.next,second.next
            first.next = second 
            second.next = temp1 
            first,second = temp1,temp2
        

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.reorderList()
ll.printList()
```

```python
#Remove Nth Node From End of List
#Med Remove nth node from the end
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None 
    
class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode, newnode
        temp = self.head 
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode 
        self.length += 1
    
    def printList(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def removeNth(self,n):
        dummy = Node(0)
        dummy.next = self.head   
        l,r = self.head,self.head.next 
        while r.next is not None and n > 0:
            l = l.next
            r = r.next
            n -= 1
        while r is not None:
            r = r.next 
        l.next = l.next.next
        return dummy.next
    
ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
new_head = ll.removeNth(2)
ll.head = new_head
ll.printList()
```
