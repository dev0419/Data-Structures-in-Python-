```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None 
    
class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode, newnode
        temp = self.head 
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode 
        self.length += 1

    def prepend(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode, newnode
        newnode.next = self.head 
        self.head = newnode 
        self.length += 1

    def printList(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def insert(self,index,value):
        if index < 0 or index > self.length:
            return False
        if index == 0:
            return self.prepend(value)
        if index == self.length:
            return self.append(value)
        newnode = Node(value)
        temp = self.head    
        for i in range(index - 1):
            temp = temp.next
        newnode.next = temp.next
        temp.next = newnode 
        self.length += 1

    def pop(self):
        if self.head is None:
            return None
        temp = self.head 
        prev = None 
        while temp.next is not None:
            prev = temp
            temp = temp.next
        prev.next = None 
        self.length -= 1
        if self.length == 0:
            self.head = None
        return temp
 

    def deleteFirst(self):
        if self.head is None:
            return None
        temp = self.head
        self.head = self.head.next
        temp.next = None
        self.length -= 1
        if self.length == 0:
            self.tail = None
        return temp
    
    def remove(self,index):
        if index < 0 or index >= self.length:
            return False 
        if index == 0:
            return self.deleteFirst()
        if index == self.length - 1:
            return self.pop()
        temp,prev = self.head,None 
        for i in range(index - 1):
            prev = temp 
            temp = temp.next
        prev.next = temp.next
        temp.next = None
        self.length -= 1 
        if self.length == 0:
            self.head,self.tail = None,None
        return temp 

    def reverse(self):
        if self.head is None:
            return None 
        temp,prev = self.head,None 
        while temp is not None:
            nxt = temp.next 
            temp.next = prev 
            prev = temp 
            temp = nxt
        return prev

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.insert(1,8)
#new_head = ll.reverse()
#ll.head = new_head 
#ll.printList()
#print(ll.pop().value)
#print(ll.deleteFirst().value)        
```

```python
#Medium  Reorder list 
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None 
    
class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode, newnode
        temp = self.head 
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode 
        self.length += 1
    
    def printList(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def reorder(self):
        slow,fast = self.head, self.head
        
        while fast and fast.next is not None:
            slow = slow.next
            fast = fast.next.next

        second = slow.next
        prev = slow.next = None 
        
        while second is not None:
            temp = second.next 
            second.next = prev 
            prev = second 
            second = temp
        
        first,second = self.head,prev
        while second is not None:
            temp1,temp2 = first.next,second.next
            first.next = second
            second.next = temp1 
            first = temp1 
            second = temp2 

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.reorder()
ll.printList()
```

```python
#Med Remove nth node from the end
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None 
    
class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode, newnode
        temp = self.head 
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode 
        self.length += 1
    
    def printList(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def removeNth(self,n):
        dummy = Node(0)
        dummy.next = self.head   
        l,r = self.head,self.head.next 
        while r.next is not None and n > 0:
            l = l.next
            r = r.next
            n -= 1
        while r is not None:
            r = r.next 
        l.next = l.next.next
        return dummy.next
    
ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
new_head = ll.removeNth(2)
ll.head = new_head
ll.printList()
```

```python
#Easy - Linked List Cycle

class Node:
    def __init__(self,value):
        self.value = value
        self.next = None 
    
class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode, newnode
        temp = self.head 
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode 
        self.length += 1
    
    def printList(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
        
    def hasCycle(self):
        slow,fast = self.head,self.head
        while fast is not None and fast.next is not None:
            slow = slow.next 
            fast = fast.next.next 
            if slow == fast:
                return True 
        return False

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.head.next = ll.tail
print(ll.hasCycle())
```
