```python
#Find the duplicate number
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode
        self.length += 1

    def printList(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    def findDuplicate(self):
        slow,fast = self.head,self.head
        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                break

        slow2 = self.head
        while slow.next is not None:
            slow = slow.next 
            slow2 = slow2.next
            
            if slow == slow2:
                break
        return slow.value

l = LinkedList(1)
l.append(3)
l.append(4)
l.append(2)
l.append(2)
print(l.findDuplicate())
```

```python
#remove the nth node from the end and return the head 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode
        self.length += 1

    def printList(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    def removeNth(self,n):
        dummy = Node(0)
        dummy.next = self.head 
        left,right = dummy,self.head 
        while right is not None and n > 0:
            right = right.next 
            n -= 1
        
        while right is not None:
            left = left.next
            right = right.next

        left.next = left.next.next 
        return dummy.next 
l = LinkedList(1)
l.append(3)
l.append(4)
l.append(2)
l.append(5)
l.head = l.removeNth(2)
l.printList()
```

```python
#Reorder Linked list 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode
        self.length += 1

    def printList(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
    
    def reorder(self):
        slow,fast = self.head,self.head.next
        while fast is not None and fast.next is not None:
            slow = slow.next 
            fast = fast.next.next 

        second = slow.next 
        prev = slow.next = None
        while second is not None:
            nxt = second.next
            second.next = prev
            prev = second 
            second = nxt

        first,second = self.head,prev
        while second is not None:
            temp1,temp2 = first.next,second.next 
            first.next = second 
            second.next = temp1 
            first,second = temp1,temp2 

l = LinkedList(1)
l.append(2)
l.append(3)
l.append(4)
l.append(5)
l.reorder()
l.printList()


        
```

```python
#Merge k sorted LinkedLists 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
        self.length = 1
    
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = newnode
        self.length += 1

    def printList(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next

    def merge_K_sorted_lists(self,lists):
        if len(lists) == 0 or not lists: 
            return None 
        #Iterate until only one ll remains which is the output
        while len(lists) > 1:
            mergedList = []
            for i in range(0, len(lists),2):
                l1 = lists[i]
                l2 = lists[i + 1] if (i + 1) < len(lists) else None
                mergedList.append(self.mergeList(l1,l2))
            lists = mergedList
        return lists[0]
        
    
    def mergeList(self,l1,l2):
        dummy = Node(0)
        dummy = self.head
        temp = dummy 
        while l1 is not None and l2 is not None:
            if l1.value < l2.value:
                temp.next = l1
                l1 = l1.next
            else:
                temp.next = l2
                l2 = l2.next
            temp = temp.next 

            if l1 is not None:
                temp.next = l1
            elif l2 is not None:
                temp.next = l2 
        return dummy.next

list1 = LinkedList(1)
list1.append(4)
list1.append(5)

list2 = LinkedList(1)
list2.append(3)
list2.append(4)

list3 = LinkedList(2)
list3.append(6)

# Merge the lists
merged_list = LinkedList(None)
result = merged_list.merge_K_sorted_lists([list1.head, list2.head, list3.head])

# Print the merged list
merged_list.head = result
merged_list.printList()
```
