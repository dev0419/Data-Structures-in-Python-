```python
#middle node 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode, newnode 
        
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode 
        self.tail.next = newnode
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def middle(self):
        slow,fast = self.head,self.head
        # the reason why fast.next is one of the conditions in the loop is when we have an even no of nodes the fast ptr will go out of bounds 
        while fast and fast.next is not None:
            slow = slow.next 
            fast = fast.next.next
        return slow


ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
mid = ll.middle()
print(f"The middle node {mid.value}")
        
        
```

```python
#linked list cycle
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode, newnode 
        
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode 
        self.tail.next = newnode
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def cycle(self):
        slow,fast = self.head,self.head
        while fast and fast.next is not None:
            slow = slow.next 
            fast = fast.next.next 
            if slow  ==  fast:
                return True
        return False

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.tail.next = ll.head
ll.cycle() 
```

```python
#Removing the Nth node
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode, newnode 
        
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode 
        self.tail.next = newnode
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def removeNth(self,n):
        dummy = Node(0)
        dummy.next = self.head 
        left,right = dummy,self.head
        while n > 0 and right  is not None:
            right = right.next 
            n -= 1
        #iteration continues until right reaches the end of the linked list
        while right is not None:
            left = left.next 
            right = right.next 
        
        ans = left.next
        left.next  = left.next.next
        return ans

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)

n = ll.removeNth(2)
print(n.value)



```

```python
#reverse a sublist within a linked list 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None 

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode, newnode 
        
    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head,self.tail = newnode,newnode 
        self.tail.next = newnode
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def reverse(self,left,right):
        dummy = Node(0)
        dummy.next = self.head 
        leftPrev,curr = dummy,self.head
        #bring the pointers to the correct position 
        for _ in range(left - 1):
            leftPrev = curr 
            curr = curr.next 
        
        #Reverse the list 
        prev = None 
        for _ in range(right - left + 1):
            tempNext = curr.next
            curr.next = prev
            #ptrs move forward
            prev,curr = curr, tempNext
        
        #Adjust pointers 
        leftPrev.next.next = curr 
        leftPrev.next = prev

        return dummy.next 

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.reverse(2,4)
ll.print_list()


            
         
#Be careful with leftPrev and prev just dont consider the both as same variable

    
```
