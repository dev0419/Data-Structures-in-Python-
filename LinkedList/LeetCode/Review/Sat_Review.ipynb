```python
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode  

    def reorder(self):
        slow,fast = self.head, self.head.next

        while fast and fast.next is not None:
            slow = slow.next 
            fast = fast.next.next 
        
        second = slow.next 
        slow.next,prev = None,None
        
        while second is not None:
            temp = second.next 
            second.next = prev 
            prev = second
            second = temp 
        
        first,second = self.head,prev 
        while second is not None:
            temp1,temp2 = first.next,second.next 
            first.next = second 
            second.next = temp1 
            first = temp1 
            second = temp2 

    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head = newnode
            self.tail = newnode
        self.tail.next = newnode
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next 

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.reorder()
ll.print_list()

        
```

```python
#Reverse linked list 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode  

    def reverse(self):
        curr = self.head 
        prev = None 
        while curr is not None:
            temp = curr.next 
            curr.next = prev 
            prev = curr 
            curr = temp
        return prev


    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head = newnode
            self.tail = newnode
        self.tail.next = newnode
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next 

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
new_head = ll.reverse()
ll.head = new_head 
ll.print_list()

        
```

```python
#Add Two Numbers 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode  

    def append(self,value):
        newnode = Node(value)
        if self.head is None:
            self.head = newnode
            self.tail = newnode
        self.tail.next = newnode
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next 
    
    def addTwoNum(self,l1,l2):
        dummy = Node(0)
        curr = dummy
        carry = 0 
        while l1 or l2 is not None:
            v1 = l1.value if l1 else 0
            v2 = l2.value if l2 else 0
            sum = v1 + v2 + carry 

            carry = sum // 10 #extract the carry
            sum = sum % 10 #keep the ones place
            curr.next = Node(sum)

            curr = curr.next 
            l1 = l1.next if l1 else None 
            l2 = l2.next if l2 else None
        
        return dummy.next
    
l1 = LinkedList(1)
l1.append(2)
l1.append(3)

l2 = LinkedList(9)
l2.append(4)
l2.append(5)

ll = LinkedList(0)
new_head = ll.addTwoNum(l1.head,l2.head)
ll.head = new_head
ll.print_list()
```
