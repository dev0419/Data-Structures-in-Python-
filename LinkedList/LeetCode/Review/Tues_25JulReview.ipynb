```python
#DEEPCOPY means for every node thats copied memory needs to be allocated
class Node:
    def __init__(self,value,random=None,next=None):
        self.value = value 
        self.random = random
        self.next = next

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
    
    def copyToRandom(self):
        curr = self.head 
        oldToCopy = {None:None} #Hashmap

        while curr is not None:
            copy = Node(curr.value)
            oldToCopy[curr] = copy
            curr = curr.next

        curr = self.head 
        
        while curr is not None:
            copy = oldToCopy[curr]
            copy.next = oldToCopy[curr.next]
            copy.random = oldToCopy[curr.random]
            curr = curr.next 

        return oldToCopy[self.head]
    
    
    def append(self, value):
        newnode = Node(value)
        if self.head is None:
            self.head, self.tail = newnode, newnode 
        self.tail.next = newnode 
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next

ll = LinkedList(7)
ll.append(13)
ll.append(11)
ll.append(10)
ll.append(1)
new_head = ll.copyToRandom()
ll.head = new_head
ll.print_list()
```

```python
#Reorder Linked List 

class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
    
    def append(self, value):
        newnode = Node(value)
        if self.head is None:
            self.head, self.tail = newnode, newnode 
        self.tail.next = newnode 
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next

    def reorder(self):
        slow,fast = self.head,self.head.next
        while fast and fast.next is not None:
            slow = slow.next
            fast = fast.next.next

        second = slow.next
        slow.next = None
        prev = None

        while second is not None:
            temp = second.next 
            second.next = prev 
            prev = second 
            second = temp

        first,second = self.head,prev 

        while second is not None:
            temp1,temp2 = first.next,second.next
            first.next = second 
            second.next = temp1
            first = temp1 
            second = temp2
        

ll = LinkedList(7)
ll.append(13)
ll.append(11)
ll.append(10)
ll.append(1)
ll.reorder()
ll.print_list()
```

```python
#AddTwoNumbers 

class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
    
    def append(self, value):
        newnode = Node(value)
        if self.head is None:
            self.head, self.tail = newnode, newnode 
        self.tail.next = newnode 
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next

def addTwoNum(l1,l2):
    dummy = Node(0)
    curr = dummy 
    carry = 0
    while l1 is not None or l2 is not None or carry != 0:
        v1 = l1.value if l1 else None
        v2 = l2.value if l2 else None
        val =  v1 + v2 + carry 
        carry = val // 10
        val = val%10
        
        curr.next = Node(val)
        curr = curr.next 
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None
    return dummy.next

l1 = LinkedList(1)
l1.append(2)
l1.append(3)

l2 = LinkedList(4)
l2.append(5)
l2.append(6)

new_head = addTwoNum(l1.head,l2.head)
ll = LinkedList(0)
ll.head = new_head
ll.print_list()
```

```python
#Remove nth node 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
    
    def append(self, value):
        newnode = Node(value)
        if self.head is None:
            self.head, self.tail = newnode, newnode 
        self.tail.next = newnode 
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next

    def removeNth(self,n):
        dummy = Node(0)
        dummy.next = self.head 
        slow,fast = dummy ,self.head 
        while n > 0 and fast is not None:
            fast = fast.next 
            n -= 1
        
        while fast is not None:
            slow = slow.next 
            fast = fast.next 
        
        slow.next = slow.next.next 
        return dummy.next
    
ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
new_head = ll.removeNth(2)
ll.head = new_head
ll.print_list()
```

```python
#Reverse a sublist within a linked list
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
    
    def append(self, value):
        newnode = Node(value)
        if self.head is None:
            self.head, self.tail = newnode, newnode 
        self.tail.next = newnode 
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next

    def reverse(self,left,right):
        dummy = Node(0)
        dummy.next = self.head 
        leftPrev,curr = dummy, self.head 
        for i in range(left-1):
            leftPrev,curr = curr,curr.next 
        
        prev = None
        for i in range(right - left + 1):
            temp = curr.next 
            curr.next = prev 
            prev = curr 
            curr = temp 
        
        leftPrev.next.next = curr 
        leftPrev.next = prev
        return dummy.next 

ll = LinkedList(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
new_head = ll.reverse(2,4)
ll.head = new_head
ll.print_list()
```

```python
#Partition List 
class Node:
    def __init__(self,value):
        self.value = value 
        self.next = None

class LinkedList:
    def __init__(self,value):
        newnode = Node(value)
        self.head,self.tail = newnode,newnode
    
    def append(self, value):
        newnode = Node(value)
        if self.head is None:
            self.head, self.tail = newnode, newnode 
        self.tail.next = newnode 
        self.tail = newnode 
    
    def print_list(self):
        temp = self.head 
        while temp is not None:
            print(temp.value)
            temp = temp.next

    def partition(self,n):
        left,right = Node(0),Node(0)
        ltail,rtail = left,right 
        while self.head:
            if self.head.value < n: 
                ltail.next = self.head 
                ltail = ltail.next 
            else:
                rtail.next = self.head
                rtail = rtail.next 
            self.head = self.head.next
        ltail.next = right
        rtail.next = None
        return left.next 
    
            

ll = LinkedList(1)
ll.append(4)
ll.append(3)
ll.append(2)
ll.append(5)
ll.append(2)
new_head = ll.partition(3)
ll.head = new_head
ll.print_list()
```
